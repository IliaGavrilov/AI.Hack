```python
import pandas as pd
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('RFM_big.csv')
df.head()
```

```python
df.tail()
```

```python
from scipy.cluster import hierarchy
def hierarchy_draw(Z, level):
    """Рисуем дендрограмму и сохраняем её"""
    plt.figure(figsize=(20,20))
    hierarchy.dendrogram(Z, color_threshold=level, leaf_font_size=5, count_sort=True)
    plt.show()
```

```python
y = df['client_id']
y = y
df = df.drop(['client_id','max','last_day'],axis=1)

df = (df - df.mean()) / (df.max() - df.min())

df.head(50)
```

```python

metrics = ['chebyshev', 'cityblock', 'correlation', 'cosine', 'euclidean']
for metric in metrics:
    print(metric)
    dist = pdist(df, metric)
    plt.hist(dist, 500, color='green', alpha= 0.5)
    plt.show()
    
    Z_cos = hierarchy.linkage( dist, method='average' )
    
    hierarchy_draw(Z_cos, 0.35)
```
